```
import efficientnet.keras as efn 
from iterstrat.ml_stratifiers import MultilabelStratifiedShuffleSplit
```

```
conv_base =  efn.EfficientNetB6(weights='imagenet',
                  include_top=False,
                  input_shape=(150, 150, 3))
```

```
from tensorflow.keras.applications import InResNetV2

conv_base = InResNetV2(weights='imagenet',
                  include_top=False,
                  input_shape=(150, 150, 3))
```

```
conv_base.summary()
```

```
import os 
import numpy as np 

from tensorflow.keras.preprocessing.image import ImageDataGenerator

base_dir = '/content/subset'
train_dir = os.path.join(base_dir, 'train') 
validation_dir = os.path.join(base_dir, 'validation') 

datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=40,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode='nearest') 

batch_size = 20

def extract_features(directory, sample_count):
  features = np.zeros(shape=(sample_count, 5, 5, 2304))
  labels = np.zeros(shape=(sample_count))
  generator = datagen.flow_from_directory(directory,
                                          target_size=(150, 150),
                                          batch_size=batch_size,
                                          class_mode='sparse')
  i=0
  for inputs_batch, labels_batch in generator:
    features_batch = conv_base.predict(inputs_batch)
    features[i*batch_size: (i+1)*batch_size] = features_batch
    labels[i*batch_size: (i+1)*batch_size] = labels_batch
    i += 1
    if i*batch_size >= sample_count:
      break
  return features, labels

train_features, train_labels = extract_features(train_dir, 1967)
train_features = np.reshape(train_features, (1967, 5, 5, 2304))
```

```
datagen1 = ImageDataGenerator(rescale=1./255) 

def extract_features1(directory, sample_count):
  features = np.zeros(shape=(sample_count, 5, 5, 2304))
  labels = np.zeros(shape=(sample_count))
  generator = datagen1.flow_from_directory(directory,
                                          target_size=(150, 150),
                                          batch_size=batch_size,
                                          class_mode='sparse')
  i=0
  for inputs_batch, labels_batch in generator:
    features_batch = conv_base.predict(inputs_batch)
    features[i*batch_size: (i+1)*batch_size] = features_batch
    labels[i*batch_size: (i+1)*batch_size] = labels_batch
    i += 1
    if i*batch_size >= sample_count:
      break
  return features, labels

validation_features, validation_labels = extract_features1(validation_dir, 533)
validation_features = np.reshape(validation_features, (533, 5, 5, 2304))
```

```
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers

model = models.Sequential()
model.add(layers.Flatten())  
model.add(layers.Dense(128, activation='relu')) 
  
model.add(layers.Dense(40, activation='softmax'))

model.compile(optimizer='adam', ##hyperparameter
              loss='sparse_categorical_crossentropy', 
              metrics=['acc'])

history = model.fit(train_features, 
                    train_labels,
                    epochs=100, 
                    batch_size=20, validation_data=(validation_features, validation_labels)
                    )
```
